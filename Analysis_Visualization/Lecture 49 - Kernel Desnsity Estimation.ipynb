Kernel density estimation
The KDE are constructed from Gaussian Distribution Plots on every data point and then sum them up 

KDE - Kernel Density Estimation
https://en.wikipedia.org/wiki/Kernel_density_estimation


Kernel (statistics) # other functions

Cumulative Distribution Function
https://en.wikipedia.org/wiki/Cumulative_distribution_function

```python
#Standard
import numpy as np
from numpy.random import randn
import pandas as pd
#Stats
from scipy import stats

#Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that the plots appear in the iPython Notebook
%matplotlib inline
```

```python
# doing it manually, rug plot, it put a tick where every values occurs
dataset = randn(25)
```

```python
sns.rugplot(dataset)
```

```python
# counting the tick on each bin and set the bar that high
plt.hist(dataset, alpha=0.3)
sns.rugplot(dataset)
```

```python
# KDE plot is going to represent each of these ticks with 
# a gaussian basis fucntion
sns.rugplot(dataset)

x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min, x_max, 100)

# setting the bandwidth estimation
bandwidth = ((4*dataset.std()**5) / (3*len(dataset))) ** 0.2

# list for the kernels
kernel_list = []

for data_point in dataset:
    # create a kernel for each point and append it to the kernel_list
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    # Scale for plottting
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis, kernel, color='grey', alpha=0.5)
```

```python
# to get the kernel estimation we need to sum all the previous 
sum_of_kde = np.sum(kernel_list, axis=0)
fig = plt.plot(x_axis, sum_of_kde, color='indianred')
sns.rugplot(dataset)
plt.yticks([])

plt.suptitle('Sum of the basis functions')
```

```python
# Now that manually is finished let's do it
# using just one step with seaborn
sns.kdeplot(dataset)
```

```python
# the bandwidth can be adjusted of the seaborn kde plot 
# to make it more or less into high frequency datasets
sns.rugplot(dataset, color='black')
for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(dataset, bw=bw, lw=1.8, label=bw)
```

```python
# we can choose different types of kernels
kernel_options = ['biw','cos','epa','gau','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset, kernel=kern, label=kern, shade=True)
```

```python
 # plot on the vertical axis on the y axis
sns.kdeplot(dataset, vertical=True)
```

```python
# Cumulative distribution fucntion
# describe the probability that the variable x is going to be found in that 
# distributions
sns.kdeplot(dataset, cumulative=True)
```

```python
# plot for multidimension data
mean = [0,0]
cov = [[1,0],[0,100]]
dataset2 = np.random.multivariate_normal(mean, cov, 1000)

dframe = pd.DataFrame(dataset2, columns=['X','Y'])

sns.kdeplot(dframe)
```

```python
sns.kdeplot(dframe.X, dframe.Y, shade=True)
```

```python
# specify bandwidths
sns.kdeplot(dframe.X, dframe.Y, bw=1)
```

```python
# using silverman estimation
sns.kdeplot(dframe.X, dframe.Y, bw='silverman')
```

```python
# kde joint plot
sns.jointplot('X','Y',dframe, kind='kde')
```
