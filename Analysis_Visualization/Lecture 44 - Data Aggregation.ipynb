data aggregations,methods or functions called that allow to have a scalar result afterwards
like mean, sum, count

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
```

```python
dframe = pd.read_csv('winequality-red.csv', sep=';')
```

```python
dframe.head()
```

```python
dframe['alcohol'].mean()
```

```python
def max_to_min(arr):
    return arr.max() - arr.min()
```

```python
wino = dframe.groupby('quality')
```

```python
wino.describe()
```

```python
wino.agg(max_to_min)
```

```python
wino.agg('mean')
```
