```python
# standard
import numpy as np
from numpy.random import randn
import pandas as pd

from pandas import Series, DataFrame

# stats
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# inline command
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
# combines kde and histogram
sns.distplot(dataset, bins=25)
```

```python
# getting other plots from displot 
sns.distplot(dataset, bins=25, rug=True, hist=False)
```

```python
# control plots inside distplot
# by passing dicts
sns.distplot(dataset, bins=25,
           kde_kws={'color':'indianred', 'label':'KDE PLOT'}, 
           hist_kws={'color':'blue', 'label':'HIST'})
```

```python
# seaborns works really well with pandas

ser1 = Series(dataset, name='My_data')
ser1.head()
```

```python
sns.distplot(ser1, bins=25)
```
