```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Violin Plots
https://en.wikipedia.org/wiki/Violin_plot


Blox Plots
https://en.wikipedia.org/wiki/Box_plot

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
# include outliers by changing the whiskers
sns.boxplot(data = [data1,data2], whis=np.inf)
```

```python
sns.boxplot(data1, orient='vert')
```

```python
# violin plots combines box plot with kde plot

# Normal dist
data1 = stats.norm(0,5).rvs(100)

# two gamma dist, concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, 
                        -1*stats.gamma(5).rvs(50)])

# box plot data1 and data2
sns.boxplot(data = [data1,data2], whis=np.inf, orient='vert')
```

```python
# comning kde and box plots
sns.violinplot(data=[data1,data2])
```

```python
# changing the bandwidth used for the density fit
sns.violinplot(data2, bw=0.1)
```

```python
# include the individual points
# stick basically brings the idea of a rugplot to a violin plot
sns.violinplot(data1, inner='stick')
```
