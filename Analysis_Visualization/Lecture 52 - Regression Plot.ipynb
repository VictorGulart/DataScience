```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
# estimated liner progression
```

```python
# passing args
sns.lmplot('total_bill', 'tip', tips,
          scatter_kws={'marker':'o', 'color':'indianred'}, 
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
# multiple higher orders, for some quadratic fits 
# passing args
sns.lmplot('total_bill', 'tip', tips, order=4,
          scatter_kws={'marker':'o', 'color':'indianred'}, 
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
# can just be a scatter plot and not a regression
sns.lmplot('total_bill','tip',tips, fit_reg=False)
```

```python
# lmplot also works on discrete variables such as percentages
tips['pct'] = 100*(tips['tip'] / tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','pct',tips)
```

```python
# adding Jitter
sns.lmplot('size','pct',tips,x_jitter=.1)
```

```python
# tendency
sns.lmplot('size','pct', tips, x_estimator=np.mean)
```

```python
# automatically define subsets along the column
sns.lmplot('total_bill','pct',tips, hue='sex', markers=['x','o'])
```

```python
sns.lmplot('total_bill', 'pct', tips, hue='day')
```

```python
# seaborn model regression fitting
sns.lmplot('total_bill', 'pct', tips, lowess=True, line_kws={'color':'black'})
```

```python
# lmplot is using a lower level function regplot
sns.regplot('total_bill', 'pct', tips)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2, sharey=True)
sns.regplot('total_bill', 'pct', tips, ax=axis1)
sns.violinplot(x='size', y='pct', data=tips, scale='width', ax=axis2)
```
