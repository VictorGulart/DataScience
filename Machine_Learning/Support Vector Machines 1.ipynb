```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data
```

```python
y = iris.target
```

```python
# for description
# iris.DESCR
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the data for training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=3)
```

```python
model.fit(x_train, y_train)
```

```python
predicted = model.predict(x_test)
expected = y_test
```

```python
print(metrics.accuracy_score(expected, predicted))
```

```python
from sklearn import svm
```

```python
x = iris.data[:,:2] # selecting just the first 2 features
```

```python
y = iris.target
```

```python
# setting the SVM regularization parameter
c = 1.0
```

```python
svc = svm.SVC(kernel='linear', C=c).fit(x,y)
```

```python
#gaussian radialbasis function
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=c).fit(x,y)
```

```python
# 3rd degree SVC polynomial
poly_svc = svm.SVC(kernel='poly', degree=3, C=c).fit(x,y)
```

```python
# linear SVC but is a one vs all
lin_svc = svm.LinearSVC(C=c).fit(x,y)
```

```python
# define a mash to plot and define a max and min
# based on the smallest and largest features in the dataset
# numpy has a nice mesh grid method

h = 0.02 # step size

x_min = x[:,0].min() - 1
x_max = x[:,0].max() + 1
```

```python
y_min = x[:,1].min() - 1
y_max = x[:,1].max() + 1
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
titles = ['SVC w/ Linear Kernel',
         'LinearSVC (linear kernel)',
         'SVC with RBF kernel',
         'SVC with polynomial (degree 3) kernel']
```

```python
 for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
        plt.figure(figsize=(20,20))
        plt.subplot(2,2, i+1)
        plt.subplots_adjust(wspace=0.3, hspace=0.3) # space between graphs
        z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        z = z.reshape(xx.shape)
        
        plt.contourf(xx,yy,z, cmap=plt.cm.terrain, alpha=0.5)
        plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.Dark2)
        plt.xlabel('Sepal Length')
        plt.ylabel('Sepal Width')
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.xticks(())
        plt.yticks(())
        plt.title(titles[i])
```
